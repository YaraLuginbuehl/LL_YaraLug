```python
import os
import matplotlib.pyplot as plt
import numpy as np

from ReadOut import read_bin_file

from datetime import timedelta
import matplotlib.dates as mdates
from matplotlib.ticker import FuncFormatter
```

```python

test1 = read_bin_file("Data/Drop_Test_2605.bin")
time = np.arange(len(test1['accelX'])) * 1 / 40  # in seconds


def seconds_to_hms(x, pos):
    hours = int(x // 3600)
    minutes = int((x % 3600) // 60)
    seconds = int(x % 60)
    return f'{hours:02}:{minutes:02}:{seconds:02}'

hms_formatter = FuncFormatter(seconds_to_hms)

# Accel X
fig, ax = plt.subplots(figsize=(30, 4))
ax.plot(time, test1['accelX'], "k", label=f'STD = {test1["accelX"].std():.2f}, mean: {test1["accelX"].mean():.2f}', markersize=1)
ax.xaxis.set_major_formatter(hms_formatter)
fig.autofmt_xdate()
ax.set_title("Acceleration X [m/s²]")
ax.set_xlabel("Time (HH:MM:SS)")
ax.set_ylabel("Acceleration")
ax.legend()

# Accel Y
fig, ax = plt.subplots(figsize=(30, 4))
ax.plot(time, test1['accelY'], "k", label=f'STD = {test1["accelY"].std():.2f}, mean: {test1["accelY"].mean():.2f}', markersize=1)
ax.xaxis.set_major_formatter(hms_formatter)
fig.autofmt_xdate()
ax.set_title("Acceleration Y [m/s²]")
ax.set_xlabel("Time (HH:MM:SS)")
ax.set_ylabel("Acceleration")
ax.legend()

# Accel Z
fig, ax = plt.subplots(figsize=(30, 4))
ax.plot(time, test1['accelZ'], "k", label=f'STD = {test1["accelZ"].std():.2f}, mean: {test1["accelZ"].mean():.2f}', markersize=1)
ax.xaxis.set_major_formatter(hms_formatter)
fig.autofmt_xdate()
ax.set_title("Acceleration Z [m/s²]")
ax.set_xlabel("Time (HH:MM:SS)")
ax.set_ylabel("Acceleration")
ax.legend()

plt.show()
```
