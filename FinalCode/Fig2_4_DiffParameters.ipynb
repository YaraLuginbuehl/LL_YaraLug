```python
from scipy.interpolate import interp1d
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import detrend
import numpy as np
```

```python

# Load the CSV
df = pd.read_csv("Data/DifferentRadii.csv", header=0, delimiter=",", quotechar='"')

label1 = 'Simulation 1 x [R = 50m, d = 45m, Regolith = 2m]'
label1g =  'Simulation 1 gcorr [R = 50m, d = 45m, Regolith = 2m]'
x1 = df[label1].values
gcorr1 = df[label1g].values

label2 = 'Simulation 2 x [R = 100m, d = 45m, Regolith = 2m]'
label2g = 'Simulation 2 gcorr [R = 100m, d = 45m, Regolith = 2m]'
x2 = df[label2].values
gcorr2 = df[label2g].values

label3 = 'Simulation 3 x [R = 150m, d = 45m, Regolith = 2m]'
label3g = 'Simulation 3 gcorr [R = 150m, d = 45m, Regolith = 2m]'
x3 = df[label3].values
gcorr3 = df[label3g].values

label4 = 'Simulation 4 x [R = 200m, d = 45m, Regolith = 2m]'
label4g = 'Simulation 4 gcorr [R = 200m, d = 45m, Regolith = 2m]'
x4 = df[label4].values
gcorr4 = df[label4g].values

label5 = 'Simulation 5 x [R = 250m, d = 45m, Regolith = 2m]'
label5g = 'Simulation 5 gcorr [R = 250m, d = 45m, Regolith = 2m]'
x5 = df[label5].values
gcorr5 = df[label5g].values



```

```python
# Load the CSV
df = pd.read_csv("Data/DifferentDepths.csv", header=0, delimiter=",", quotechar='"')

label6 = 'Simulation 1 x [R = 75m, d = 25m, Regolith = 2m]'
label6g = 'Simulation 1 gcorr [R = 75m, d = 25m, Regolith = 2m]'
x6 = df[label6].values
gcorr6 = df[label6g].values

label7 = 'Simulation 2 x [R = 75m, d = 30m, Regolith = 2m]'
label7g = 'Simulation 2 gcorr [R = 75m, d = 30m, Regolith = 2m]'
x7 = df[label7].values
gcorr7 = df[label7g].values

label8 = 'Simulation 3 x [R = 75m, d = 35m, Regolith = 2m]'
label8g = 'Simulation 3 gcorr [R = 75m, d = 35m, Regolith = 2m]'
x8 = df[label8].values
gcorr8 = df[label8g].values

label9 = 'Simulation 4 x [R = 75m, d = 40m, Regolith = 2m]'
label9g = 'Simulation 4 gcorr [R = 75m, d = 40m, Regolith = 2m]'
x9 = df[label9].values
gcorr9 = df[label9g].values

label10 = 'Simulation 5 x [R = 75m, d = 45m, Regolith = 2m]'
label10g = 'Simulation 5 gcorr [R = 75m, d = 45m, Regolith = 2m]'
x10 = df[label10].values
gcorr10 = df[label10g].values
```

```python
df = pd.read_csv("Data/DifferentHeights.csv", header=0, delimiter=",", quotechar='"')

labelh1 = 'Simulation 1 x [R = 75m, d = 25m, Regolith = 2m, H = 10m]'
labelh1g = 'Simulation 1 gcorr [R = 75m, d = 25m, Regolith = 2m, H = 10m]'
xh1 = df[labelh1].values
gcorrh1 = df[labelh1g].values

labelh2 = 'Simulation 2 x [R = 75m, d = 30m, Regolith = 2m, H = 20m]'
labelh2g = 'Simulation 2 gcorr [R = 75m, d = 30m, Regolith = 2m, H = 20m]'
xh2 = df[labelh2].values
gcorrh2 = df[labelh2g].values

labelh3 = 'Simulation 3 x [R = 75m, d = 35m, Regolith = 2m, H = 30m]'
labelh3g = 'Simulation 3 gcorr [R = 75m, d = 35m, Regolith = 2m, H = 30m]'
xh3  = df[labelh3].values
gcorrh3  = df[labelh3g].values

labelh4 = 'Simulation 4 x [R = 75m, d = 40m, Regolith = 2m, H = 40m]'
labelh4g = 'Simulation 4 gcorr [R = 75m, d = 40m, Regolith = 2m, H = 40m]'
xh4 = df[labelh4].values
gcorrh4 = df[labelh4g].values

labelh5 = 'Simulation 5 x [R = 75m, d = 45m, Regolith = 2m, H = 50m]'
labelh5g = 'Simulation 5 gcorr [R = 75m, d = 45m, Regolith = 2m, H = 50m]'
xh5 = df[labelh5].values
gcorrh5 = df[labelh5g].values
```

```python
df = pd.read_csv('Data/Baseline.csv')

mask = (df['x'] >= x1.min()) & (df['x'] <= x1.max())

x = df['x'].values[mask]
g = df['g'].values[mask]



f = interp1d(x,g, kind='linear', fill_value='extrapolate')
g_new = f(x1)
```

```python

plt.rcParams.update({
    "font.size": 17,            # base font size
    "axes.titlesize": 17,       # title font
    "axes.labelsize": 17,       # axis labels
    "xtick.labelsize": 17,
    "ytick.labelsize": 17,
    "legend.fontsize": 17,
})

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 16), sharex=True)


# ax1.axvline(x=-300, color='red', linestyle='--')
# ax1.axvline(x=300, color='red', linestyle='--')
# ax2.axvline(x=-300, color='red', linestyle='--')
# ax2.axvline(x=300, color='red', linestyle='--')
# ax3.axvline(x=-300, color='red', linestyle='--')
# ax3.axvline(x=300, color='red', linestyle='--')


x_values = [x1, x2, x3, x4, x5]
gcorr_values = [gcorr1, gcorr2, gcorr3, gcorr4, gcorr5]
labels = ['R = 50m', 'R = 100m', 'R = 150m', 'R = 200m', 'R = 250m']

for x, gcorr, label in zip(x_values, gcorr_values, labels):
    detrended = gcorr-g_new
    
    # Find the index closest to x = 300
    idx_600 = np.argmin(np.abs(x - 600))
    idx_m600 = np.argmin(np.abs(x + 600))
    value_at_600 = (detrended[idx_600] + detrended[idx_m600])/2
    detrended2 = detrended - value_at_600
    min_value = np.min(detrended)
    diff = min_value

    print(f"{label}: Value at x=600 = {value_at_600:.3f}, Min = {min_value:.3f}, Difference = {diff:.3f}")

    ax1.plot(x, detrended2, label=None, color="Cornflowerblue")


ax1.set_xlim(-500, 500)
ax1.set_title(f'Varying Radii (d = 45m, H = 30m)')
ax1.set_ylabel("g [mGal]")
ax1.grid()


x_values = [x6, x7, x8, x9, x10]
gcorr_values = [gcorr6, gcorr7, gcorr8, gcorr9, gcorr10]
labels = ["d = 25m","d = 30m","d = 35m","d = 40m","d = 45m"]

for x, gcorr, label in zip(x_values, gcorr_values, labels):
    detrended = gcorr-g_new
    
    # Find the index closest to x = 300
    idx_600 = np.argmin(np.abs(x - 600))
    idx_m600 = np.argmin(np.abs(x + 600))
    value_at_600 = (detrended[idx_600] + detrended[idx_m600])/2
    detrended2 = detrended - value_at_600
    min_value = np.min(detrended)
    diff = min_value

    print(f"{label}: Value at x=600 = {value_at_600:.3f}, Min = {min_value:.3f}, Difference = {diff:.3f}")

    ax2.plot(x, detrended2, label=label, color="Cornflowerblue")

ax2.set_xlim(-500, 500)
ax2.set_ylim(-4,1)
ax2.set_title(f'Varying Depths (R = 75m, H = 30m)')
ax2.set_ylabel("g [mGal]")
ax2.grid()

x_values = [xh1, xh2, xh3, xh4, xh5]
gcorr_values = [gcorrh1, gcorrh2, gcorrh3, gcorrh4, gcorrh5]
labels = ["H = 10m","H = 20m","H = 30m","H = 40m","H = 50m"]

for x, gcorr, label in zip(x_values, gcorr_values, labels):
    detrended = gcorr-g_new
    
    # Find the index closest to x = 300
    idx_600 = np.argmin(np.abs(x - 600))
    idx_m600 = np.argmin(np.abs(x + 600))
    value_at_600 = (detrended[idx_600] + detrended[idx_m600])/2
    detrended2 = detrended - value_at_600
    min_value = np.min(detrended)
    diff = min_value

    print(f"{label}: Value at x=600 = {value_at_600:.3f}, Min = {min_value:.3f}, Difference = {diff:.3f}")

    ax3.plot(x, detrended2, label=label, color="Cornflowerblue")

ax3.set_xlim(-500, 500)
ax3.set_ylim(-4,1)
ax3.set_title(f'Varying Heights (R = 75m, d = 45m)')
ax3.set_ylabel("g [mGal]")
ax3.set_xlabel("x [m]")
ax3.grid()
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("Outputs/VaryingParameters.png")
plt.show()

```
