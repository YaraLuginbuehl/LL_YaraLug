```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import detrend
```

```python
# Read CSV
df = pd.read_csv('Data/Baseline.csv')

mask = (df['x'] >= -300) & (df['x'] <= 300)

x = df['x']
g = df['g']

x_sub = df['x'][mask]
g_sub = df['g'][mask]

```

```python
df = pd.read_csv('Data/Baseline_wRille.csv')

mask = (df['x'] >= -300) & (df['x'] <= 300)

x_wRille_sub = df['x'][mask]
g_wRille_sub = df['g'][mask]

x_wRille = df['x'].values
g_wRille = df['g'].values
```

```python

df = pd.read_csv('Data/Baseline_wCave.csv')

mask = (df['x'] >= -300) & (df['x'] <= 300)

x_wCave = df['x'].values
g_wCave = df['g'].values
x_wCave_sub = df['x'][mask]
g_wCave_sub = df['g'][mask]
```

```python

df = pd.read_csv('Data/Baseline_wRille_wRegolith.csv')

mask = (df['x'] >= -300) & (df['x'] <= 300)

x_wRR = df['x'].values
g_wRR = df['g'].values

x_wRR_sub = df['x'][mask]
g_wRR_sub = df['g'][mask]

```

```python


fig, axes = plt.subplots(1, 3, figsize=(14, 4), sharey=True)

plt.rcParams.update({
    "font.size": 13,         
    "axes.titlesize": 13,       
    "axes.labelsize": 13,       
    "xtick.labelsize": 13,
    "ytick.labelsize": 13,
    "legend.fontsize": 13,
})
plt.grid(True)
plt.title("Elements in Gravimetric Signal")

axes[1].plot(x, g, color='black', label='Framework')
axes[1].plot(x_wRille, g_wRille, color='cornflowerblue', label='With Rille')
# axes[1].legend()
axes[1].set_title('Rille')
axes[1].set_xlabel('x [m]')
# axes[1].set_ylabel('g [mGal]')
axes[1].set_xlim(-400,400)
axes[1].grid(True)


axes[0].plot(x, g, color='black', label='Framework')
axes[0].plot(x_wCave, g_wCave, color='cornflowerblue', label='With Tube ()')
# axes[0].legend()
axes[0].set_title('Tube (R = 75m, d = 45m)')
axes[0].set_xlabel('x [m]')
axes[0].set_ylabel('g [mGal]')
axes[1].set_ylim(27.5,31)
axes[0].set_xlim(-400,400)
axes[0].grid(True)

# Third subplot
axes[2].plot(x, g, color='black', label='Framework')
axes[2].plot(x_wRR, g_wRR, color='cornflowerblue', label='With Rille and Regolith')
# axes[2].legend()
axes[2].set_title('Rille and Regolith (d = 2m)')
axes[2].set_xlabel('x [m]')
# axes[2].set_ylabel('g [mGal]')
axes[2].set_xlim(-400,400)
axes[2].grid(True)

plt.tight_layout()

plt.savefig("Outputs/ModelEnvironments.png")
plt.show()
```
